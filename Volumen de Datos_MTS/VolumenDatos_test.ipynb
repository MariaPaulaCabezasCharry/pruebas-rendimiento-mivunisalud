```python
#Importar librerías
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import plotly.express as px
import seaborn as sns
import numpy as np


# Leer el archivo de la base de datos
df_bbdd = pd.read_csv('/content/drive/MyDrive/datasets/datos_bbdd.csv')

#Leer el archivo de los registros de Agost+
df_agost = pd.read_csv('/content/drive/MyDrive/datasets/reportes_agost.csv')
```

```python
# Comparar el número de registros de cada dataset
data_bbdd = df_bbdd.shape[0]
data_agost = df_agost.shape[0]

print(f"Volumen de datos de MIV Unisalud: {data_bbdd} registros")
print(f"Volumen de datos de Agost+: {data_agost} registros")
```

```python
# Calcular el porcentaje de diferencia del volumen de datos de los datasets
volume_difference = data_bbdd - data_agost
percentage_difference = (volume_difference / data_bbdd) * 100

# Calcular cuántas veces es más grande el dataset 1 (data_bbdd) que el dataset 2 (data_agost)
times_larger = data_bbdd / data_agost

# Imprimir resultados
print(f"El volumen de datos de MIV Unisalud es {times_larger:.2f} veces más grande que el de Agost +")
print(f"Diferencia: {volume_difference} registros")
```

```python
# Crear dataframe de volúmenes
df_volumes = pd.DataFrame({
    'Dataset': ['MIV Unisalud', 'Agost+'],
    'Volumen de datos': [data_bbdd, data_agost]
})

# Graficar la comparación
plt.figure(figsize=(8,6))
sns.barplot(x='Dataset', y='Volumen de datos', data=df_volumes, hue='Dataset', palette=["#0032A0", "#E6000A"], legend=False)

# Título y gráfico
plt.title('Comparación entre MIV Unisalud y Agost +')
plt.ylabel('Número de registros')
plt.xlabel('Dataset')
plt.show()
```
