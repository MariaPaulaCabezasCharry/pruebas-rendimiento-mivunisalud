```python
import pandas as pd

# Leer el primer archivo CSV
df = pd.read_csv('./datos_tarjeta_sd.csv')

# Leer el segundo archivo CSV
df_bbdd = pd.read_csv('./datos_bbdd.csv')
```

```python
validador = 0
# Verificar que ambas columnas tienen la misma cantidad de filas
if len(df) == len(df_bbdd):
    # Comparar los valores de las filas de ambas columnas
    for i in range(len(df)):
        if df['temp'][i] != df_bbdd['temp'][i]:
            print('*'*50)
            print(f'Valor diferente encontrado en la fila {i+1}')
            print(f'Dato en tarjeta SD: {df["temp"][i]}')
            print(f'Dato en BBDD: {df_bbdd["temp"][i]}')
            validador+=1
    if validador == 0:
      print('*'*50)
      print('Los valores de los registros son iguales')
      print('*'*50)
    else:
      print('*'*50)
      print(f'Hay {validador} incompatibilidades')
      print('*'*50)
else:
    print('Las columnas no tienen la misma cantidad de filas.')
```
